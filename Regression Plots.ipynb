```python
# importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("country_vaccinations.csv") # using the new and updated vaccine dataset
```

```python
df.head() # first 5 rows
```

```python
# create a list of each unique country
country_list = list(df['country'].unique())
```

```python
len(country_list)
```

```python
plt.figure(figsize=(20,20))

for i, country in enumerate(country_list[0:16]):
    
    #plt.figure()
    
    plt.subplot(4,4,i+1)
    
    plt.title(country)
    
    # filter dataframe to current country in loop
    country_df = df[df['country'] == country] 
    
    # sort by ascending date
    date_df = country_df[['date','daily_vaccinations']].sort_values('date',ascending=True)
    
    date_df = date_df.dropna()
    
    date_df["date"] = pd.to_numeric(date_df["date"].str.replace('-',''))
    
    sns.regplot(data=date_df,x='date',y='daily_vaccinations')
    
plt.show()
```
